```python
file_name = glob("healthcare*.csv")
df = pd.read_csv(file_name[0])
df.head()
```

```python
# 성별 "Other" 행 제거 및 인덱스 리셋
df = df.drop(index = 3116)
df = df.reset_index()
# 인덱스 칼럼 제거
df = df.drop(columns = "index")
```

```python
# 결측치 제거
df = df.dropna()
```
